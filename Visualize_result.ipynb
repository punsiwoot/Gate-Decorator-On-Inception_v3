```python
import pickle
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

# Resnet50

```python
# pretrain_resnet50_metric = open("Save/Result/save_first_90_epoch.pkl", 'rb')
pretrain_resnet50_metric  = pd.read_pickle(r"Save/Result/save_first_90_epoch.pkl")
```

```python
train_resnet50_metric = pretrain_resnet50_metric["train"]
test_resnet50_metric = pretrain_resnet50_metric['test']
```

```python
train_resnet50_loss = []
test_restnet50_loss = []
test_top1_acc_resnet50 = []
for i in range(len(train_resnet50_metric)):
    train_resnet50_loss.append(train_resnet50_metric[i]["train_loss"])
    test_restnet50_loss.append(test_resnet50_metric[i]["test_loss"])
    test_top1_acc_resnet50.append(test_resnet50_metric[i]["acc@1"])

# train_resnet50_metric[0]["train_loss"]
```

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True)
ax1.plot(train_resnet50_loss)
ax1.set_title('train loss')
ax2.plot(test_restnet50_loss)
ax2.set_title('test loss')
plt.show()
plt.figure
plt.plot(test_top1_acc_resnet50)
plt.title('top 1 acc')
plt.show()
```
